```python
import pandas as pd
import numpy as np
import glob,os
```

```python
import os
path = os.getcwd() # Check current directory's path
os.chdir('/Users/xgjdreamer/Downloads/RQ1_results/results_idle_rq1/data/nexus9/com-whatsapp/trepn')
file = glob.glob(os.path.join(path, "HT4AMJT02334_Trepn_**.csv")) 
```

```python
print(file)
dl= []
for f in file:
    dl.append(pd.read_csv(f,index_col=None))
```

```python
print(len(dl))
```

```python
print(dl[0])
```

```python
sum_idle_raw_whatsapp = [];
sum_idle_delta_whatsapp = [];
ave_idle_raw_whatsapp = [];
ave_idle_delta_whatsapp = [];
number_whatsapp = [];

for x in dl:
    sum_idle_raw_whatsapp.append(x["Battery Power* [uW] (Raw)"].sum());
    sum_idle_delta_whatsapp.append(x["Battery Power* [uW] (Delta)"].sum());
    ave_idle_raw_whatsapp.append(x["Battery Power* [uW] (Raw)"].mean());
    ave_idle_delta_whatsapp.append(x["Battery Power* [uW] (Delta)"].mean());
    number_whatsapp.append(len(x));
```

```python
print(sum_idle_raw_whatsapp);
print(ave_idle_raw_whatsapp);
print(number_whatsapp);
print(len(sum_idle_raw_whatsapp));
```

```python
from pandas import DataFrame
whatsapp_idle = {"sum_idle_raw_whatsapp": sum_idle_raw_whatsapp,
                "sum_idle_delta_whatsapp" : sum_idle_delta_whatsapp,
                "ave_idle_raw_whatsapp" : ave_idle_raw_whatsapp,
                "ave_idle_delta_whatsapp" : ave_idle_delta_whatsapp,
                "number_whatsapp" : number_whatsapp}
whatsapp_idle_data = DataFrame(whatsapp_idle)#将字典转换成为数据框
print(whatsapp_idle_data)
```

```python
whatsapp_idle_data.to_csv("/Users/xgjdreamer/Downloads/whatsapp_idle_data.csv",index = True)
```
